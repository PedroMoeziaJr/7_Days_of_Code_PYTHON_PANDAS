## Unificando as tabelas carregadas utilizando a função concat

A função pd.concat serve para unir as tabelas uma em cima da outra. O pandas detecta o nome dos campos e adiciona os dados conforme os nomes. 

A função serve para esse caso pois todas as tabelas possuem campos com nomes semelhantes.

Detalhe que por padrão ele puxa os índices. As vezes esses índices podem significar nada na tabela. 

Por isso ao final usamos o comando ignore_index = True para ignorar o índice padrão

Colocamos colchetes [] dentro de () porque trata-se de uma lista. 

Assim no exemplo a emprestimos_1 recebe os dados da emprestimos_2 embaixo e assim por diante...

Além das lista e do parâmetro ignore_index entre parenteses outros parâmetros podem ser usados.

Exemplos:

1- Apresenta duas tabelas com os seguintes dados separados:

In[66] pd.concat([emprestimo_1, emprestimos_2], keys = ['ALUNO DE GRADUAÇÃO1', 'ALUNO DE PÓS-GRADUAÇÃO']) 




```python
#unificandos os dataframes, colcheters para dizer que é uma lista, .concat é a função utilizada. 
#O parâmetro ignore_index=True garante que o índice do DataFrame combinado seja redefinido, 
#para que os índices dos DataFrames originais não se sobreponham.

emprestimos_concatenados = pd.concat([emprestimos_1,
emprestimos_2,
emprestimos_3,
emprestimos_4,
emprestimos_5,
emprestimos_6,
emprestimos_7,
emprestimos_8,
emprestimos_9,
emprestimos_10,
emprestimos_11,
emprestimos_12,
emprestimos_13,
emprestimos_14,
emprestimos_15,
emprestimos_16,
emprestimos_17,
emprestimos_18,
emprestimos_19,
emprestimos_20,
emprestimos_21], ignore_index = True)

emprestimos_concatenados
```

## Utilizando a função .merge para unir os dataframes carregados

A função pd.merge mescla as tabelas com base em um campo que possua valores em comum.

É possível passar um parâmetro usando a função on com base em uma coluna

Ele vai colocar uma tabela ao lado da outra utilizando padrão INNER.

Ou seja, só vai incluir os dados que existem nas duas tabelas ao mesmo tempo

Exemplo:

todos_emprestimos = pd.merge(emprestimos_concatenados, exemplares_acervo, on = 'codigo_barras')

Se as tabelas tivesse campos com nomes diferentes:

1-Ou renomear os campos para deixar com nomes iguais.

2-Ou usar o parâmetro = pd.merge(emprestimos_concatenados, exemplares_acervo, left_on = 'codigo_barras', right_on = 'outronome')

Também é possível usar o parâmetro só how e on.

Exemplo:

todos_emprestimos = pd.merge(emprestimos_concatenados, exemplares_acervo, on = "tipo_vinculo_usuario" how='left')


##  Diferença entre INNER, OUTER, LEFT E RIGHT

Vimos que o padrão é o INNER

INNER = vai pegar somente os dados que contém nas duas tabela.
LEFT = vai pegar todos os dados da tabela da Esquerda bem como a INNER. 
RIGHT = todos os dados da tabela da direita bem como a INNER
OUTER = todas as informações que estão em todas as tabelas. Ele não é a soma, porque ainda tem o INNER

```python
#juntando os dadaframes emprestimos_concatenados + exemplares_acervo e vamos criar um data frame todos_emprestimos
#vamos utilizar a função .merge
todos_emprestimos = emprestimos_concatenados.merge(exemplares_acervo)
todos_emprestimos
```
